```python
#Task 1
# Inventori (isi sendiri)
inventory = []

# Function
def input_data():
    name = input("Masukkan nama barang: ")
    category = input("Masukkan kategori barang: ")
    price = int(input("Masukkan harga (Rupiah): "))
    inventory.append({"name": name, "category": category, "price": price})
    print("Terima kasih, data sudah disimpan\n")

# Function buat nampilin data
def display_data():
    if len(inventory) == 0:
        print("Data belum tersedia, silahkan pilih menu 1 untuk isi data\n")
    else:
        for item in inventory:
            print(f"Nama Barang: {item['name']}, Kategori: {item['category']}, Harga: {item['price']} Rupiah\n")

# OOP
while True:
    print("PROGRAM INVENTORY MANAGEMENT")
    print("Menu:")
    print("1. Input Data")
    print("2. Menampilkan data")
    print("3. Keluar")
    choice = int(input("Pilih menu atau keluar: "))
    
    if choice == 1:
        input_data()
    elif choice == 2:
        display_data()
    elif choice == 3:
        print("Terima kasih telah menggunakan program kami.")
        break
    else:
        print("Pilihan tidak valid. Silakan coba lagi.\n")
```
